```python
import os
import numpy as np
import pandas as pd
```

```python
#1. Data Collection
```

```python
file_path = "/content/Data_science_wiki.txt"

with open(file_path,"r") as f:
  text = f.read()

print(text)
```

```python
import re

cleaned_text = re.sub(r'\[[0-9]+\]', '', text)

print(cleaned_text)
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts([cleaned_text])
```

```python
len(tokenizer.index_word)
```

```python
import re

def split_text(text):
    # Use regular expression to split the text wherever ".", ",", and "?" occur
    # The regular expression pattern matches any of these characters followed by any number of spaces
    # This ensures that punctuation marks are separated from the words
    return re.split(r'\s*[\.,?]\s*', text)

sentences = split_text(cleaned_text)
print(sentences)
```

```python
tok_sent = []
for sentence in sentences:
  tok_sent.append(tokenizer.texts_to_sequences([sentence])[0])
```

```python
input_text = []

for one_sent in tok_sent:
  for i in range(1,len(one_sent)):
    input_text.append(one_sent[:i+1])
```

```python
input_text
```

```python
max_length = max([len(i) for i in input_text])
max_length
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

padded_input = pad_sequences(input_text,maxlen = max_length,padding = "pre")
```

```python
# Making X and Y from the padded Data

X = padded_input[:,:-1]
Y = padded_input[:,-1]
```

```python
X.shape
```

```python
Y.shape
```

```python
from tensorflow.keras.utils import to_categorical
y = to_categorical(Y,num_classes=567)
y
```

```python
y.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,LSTM,Dense
model = Sequential()
```

```python
model.add(Embedding(567,150,input_length=47))
model.add(LSTM(150))
model.add(Dense(567,activation="softmax"))
```

```python
model.compile(optimizer="adam",loss="categorical_crossentropy",metrics=["accuracy"])
```

```python
model.summary()
```

```python
model.fit(X,y,epochs=100)
```

```python
#prediction
```

```python
predictive_text = "Data Science and machine learning"

for i in range(50):
  # tokenize
  tokenized_input = tokenizer.texts_to_sequences([predictive_text])[0]
  # padding
  padded_input = pad_sequences([tokenized_input],maxlen=47,padding ="pre")

  # predicting
  pos = np.argmax(model.predict(padded_input))
  pos

  for word,index in tokenizer.word_index.items():
    if index == pos:
      predictive_text = predictive_text + " " + word
      print(predictive_text)
```

```python
import time
text = "what is the fee"

for i in range(10):
  # tokenize
  token_text = tokenizer.texts_to_sequences([text])[0]
  # padding
  padded_token_text = pad_sequences([token_text], maxlen=47, padding='pre')
  # predict
  pos = np.argmax(model.predict(padded_token_text))

  for word,index in tokenizer.word_index.items():
    if index == pos:
      text = text + " " + word
      print(text)
      time.sleep(2)
```
